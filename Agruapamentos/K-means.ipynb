```python
#importar os  pacotes 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns 
```

```python
#importar dados desde os conjunto de skleran
data_set = pd.read_csv("/home/edianfranco/Dropbox/Univesidade Federal do Para/Laboratorio/Curso de Machine Learning/Data_set/Iris.csv")
```

```python
#Eliminas as classes do data-set
data_set_cluster = data_set.iloc[:, [1, 2, 3, 4]].values
```

```python
#intentar descubrir a quantidade de agrupamentos
sns.pairplot(data_set.drop(labels=["Id"],axis=1))
```

```python
#Procurar o numero optimos de agrupamentos 

wcss = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(data_set_cluster)
    wcss.append(kmeans.inertia_)
```

```python
#Vizualizar a grafica para inferir o numero de agrupamentos 
plt.plot(range(1, 10), wcss)
plt.title('Metodo do cotovelo')
plt.xlabel('Numero de  clusters')
plt.ylabel('WCSS') #Suma do erro quadro dos cluster
plt.show()
```

```python
#Implementando o algoritmos com o numero de agrupamento seleccionado
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
Resultado_kmeans = kmeans.fit_predict(data_set_cluster)
```

```python
#Vizualizar os agrupamentos
plt.scatter(data_set_cluster[Resultado_kmeans == 0, 0], data_set_cluster[Resultado_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(data_set_cluster[Resultado_kmeans == 1, 0], data_set_cluster[Resultado_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(data_set_cluster[Resultado_kmeans == 2, 0], data_set_cluster[Resultado_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

#Vizualizar centroide de cada agrupamento

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'black', label = 'Centroids')

plt.legend()
```
